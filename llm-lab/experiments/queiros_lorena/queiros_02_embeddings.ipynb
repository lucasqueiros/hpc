# Embeddings

## Preparação do ambiente

```python
import sys
import os
import torch

raiz = os.path.abspath('../../')

if raiz not in sys.path:
    sys.path.append(raiz)

# Verificando versões das libs
from importlib.metadata import version
print("torch version:", version("torch"))
print("tiktoken version:", version("tiktoken"))
```

```python
pasta = '../../data/'
arquivo = 'hamlet.txt'
path = os.path.join(pasta, arquivo)

with open (path, "r", encoding="utf-8") as f:
    corpus = f.read()

print(corpus[:300])
```

## DataLoader

```python
from utils.queiros.dataloader import create_dataloader

# hiperparametros
VOCAB_SIZE      = 50257
CONTEXT_LENGTH  = 128
EMBEDDING_DIM   = 256
BATCH_SIZE      = 8

print(f"Parâmetros na camada de embedding: {VOCAB_SIZE * EMBEDDING_DIM:,}")
```

```python
dataloader = create_dataloader(corpus, batch_size=BATCH_SIZE, max_length=CONTEXT_LENGTH,
                               stride=CONTEXT_LENGTH, shuffle=True)

inputs, targets = next(iter(dataloader))
```

## Token Embeddings

```python

token_embedding_layer = torch.nn.Embedding(VOCAB_SIZE, EMBEDDING_DIM)

print(f"Pesos: {token_embedding_layer.weight.shape}")  # (50257, 256)
print(f"Requer gradiente: {token_embedding_layer.weight.requires_grad}")
```

```python
token_embeddings = token_embedding_layer(inputs)

print(f"inputs.shape          : {inputs.shape}")
print(f"token_embeddings.shape: {token_embeddings.shape}")
```
